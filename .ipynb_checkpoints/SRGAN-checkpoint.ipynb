```python
import torch
from torchvision import transforms,datasets
import helperLayers
import loss
import model
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
train_dataset = datasets.ImageFolder(
    '../dataset/',
    transforms.Compose([
        transforms.RandomResizedCrop(96),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
    ]))
```

```python
train_dataset
```

```python
data_loader = torch.utils.data.DataLoader(train_dataset,
                                          batch_size=16,
                                          shuffle=True)
```

```python
generator = model.Generator().cuda()
discriminator = model.Discriminator().cuda()
network = loss.PerceptualLoss().cuda()
adam = torch.optim.Adam(lr=1e-4,params=generator.parameters())
adam2 = torch.optim.Adam(lr=1e-4,params=discriminator.parameters())
```

```python
!pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials


auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
last_weight_file = drive.CreateFile({'id': '1d3vGDs6MI-trNXXek3si6jTtfPMBpq0q'})
last_weight_file.GetContentFile('loaded.weights')
generator.load_state_dict(torch.load('loaded.weights'))
last_weight_file2 = drive.CreateFile({'id': '11XIRouZe7WeoxWeA5djLhWLwMH7RvfFB'}) 
last_weight_file2.GetContentFile('loaded2.weights')
discriminator.load_state_dict(torch.load('loaded2.weights'))
```

```python
inp = scipy.misc.imread('../img_002_SRF_4_LR.png')
out = generator(torch.from_numpy((inp/255).transpose((2,0,1))).cuda().type(torch.cuda.FloatTensor).unsqueeze(0))
out = out.clamp(min=-1,max=1)
out = (out+1)/2
plt.imshow(out[0].cpu().detach().numpy().transpose(1,2,0))
plt.imsave('output.jpg',out[0].cpu().detach().numpy().transpose(1,2,0))
```

```python
for epoch in range(1):
  sum_loss_generator = 0
  sum_loss_between_generator = 0
  sum_loss_discriminator = 0
  sum_loss_between_discriminator = 0
  print('=========================================================')
  for step,data in enumerate(data_loader):
    hr_images = data[0].cuda()
    lr_images = np.zeros((hr_images.shape[0],hr_images.shape[1],int(hr_images.shape[2]/4),int(hr_images.shape[3]/4)))
    for i in range(hr_images.shape[0]):
        lr_image = scipy.misc.imresize(hr_images[i].cpu().numpy(),size=0.25,interp='bicubic')/255
        lr_images[i] = lr_image.transpose((2,0,1))
    lr_images = torch.from_numpy(lr_images)
    
    generated_imgs = generator(lr_images.cuda().type(torch.cuda.FloatTensor))
    generated_probs = discriminator(generated_imgs)
    ####################################################################################
    discriminator.zero_grad()
    loss_images = (-discriminator(2*hr_images-1).log().sum() - (1-generated_probs).log().sum())/32
    loss_images.backward(retain_graph=True)
    adam2.step()
    
    sum_loss_discriminator = sum_loss_discriminator + loss_images.cpu().detach().numpy()
    sum_loss_between_discriminator = sum_loss_between_discriminator + loss_images.cpu().detach().numpy()
    if step%50 == 49:
      print("Epoch Number : "+str(epoch+2)+" Step Number : "+str(step+1)+" Loss Discriminator: "+str(sum_loss_between_discriminator/50))
      sum_loss_between_discriminator = 0
    ####################################################################################
    generator.zero_grad()
    loss_images = network(generated_imgs,2*hr_images-1) - generated_probs.log().sum()*1e-3/16
    loss_images.backward(retain_graph=True)
    adam.step()
    
    sum_loss_generator = sum_loss_generator + loss_images.cpu().detach().numpy()
    sum_loss_between_generator = sum_loss_between_generator + loss_images.cpu().detach().numpy()
    if step%50 == 49:
      print("Epoch Number : "+str(epoch+2)+" Step Number : "+str(step+1)+" Loss Generator: "+str(sum_loss_between_generator/50))
      sum_loss_between_generator = 0
  print('==========================================================')
  print('Epoch '+str(epoch+2)+' completed. Loss Generator : '+str(sum_loss_generator/len(data_loader)))
  print('Epoch '+str(epoch+2)+' completed. Loss Discriminator : '+str(sum_loss_discriminator/len(data_loader)))
  file_name = 'generator'+str(epoch+2)+'_1.weights'
  torch.save(generator.state_dict(),file_name)
  file_name2 = 'discriminator'+str(epoch+2)+'_1.weights'
  torch.save(discriminator.state_dict(),file_name2)
  auth.authenticate_user()
  gauth = GoogleAuth()
  gauth.credentials = GoogleCredentials.get_application_default()
  drive = GoogleDrive(gauth)
  weights_file = drive.CreateFile({'title' : file_name})
  weights_file.SetContentFile(file_name)
  weights_file.Upload()
  weights_file = drive.CreateFile({'title' : file_name2})
  weights_file.SetContentFile(file_name2)
  weights_file.Upload()
  print('Weights Uploaded')
  inp = scipy.misc.imread('../img_002_SRF_4_LR.png')
  out = generator(torch.from_numpy((inp/255).transpose((2,0,1))).cuda().type(torch.cuda.FloatTensor).unsqueeze(0))
  out = out.clamp(min=-1,max=1)
  out = (out+1)/2
  plt.imshow(out[0].cpu().detach().numpy().transpose(1,2,0))
  plt.imsave('output.jpg',out[0].cpu().detach().numpy().transpose(1,2,0))
```
